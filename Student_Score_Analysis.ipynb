```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("students_scores.csv")
print(df.head())
```

# Data Transformation

```python
#Drop Unnamed column
df = df.drop("Unnamed: 0", axis = 1)
df.head()
```

```python
#Change weekly study hours column
df["WklyStudyHours"] = df["WklyStudyHours"].str.replace("5 - 10","5 to 10")
df.head()
```

```python
#We can keep it as 5-10 as well
df["WklyStudyHours"] = df["WklyStudyHours"].str.replace("5 to 10","5 - 10")
df.head()
```

# Data Analysis

```python
#Gender Distribution(using count plot:gives the count of female and male.)
plt.figure(figsize=(4,5)) #to define the size of the chart below by passing width and height.)
ax = sns.countplot(data =df, x = "Gender")
ax.bar_label(ax.containers[0]) # to give the total count of each as a label
plt.title("Gender Distribution", fontsize=16, fontweight='bold')
#plt.grid(True, which='both', linestyle='--', linewidth=0.5, color='gray')
plt.style.use('ggplot') 
plt.show()
```

```python
#From the above chart we have analysed that:
#Number of females in the data is more than the number of males.
```

```python
#To know how parents education affects the scores of the students
#we will use groupby fucntion.

gb = df.groupby("ParentEduc").agg({"MathScore":'mean', "ReadingScore":'mean',"WritingScore":'mean'})
gb
```

```python
#To analysis the above data using a heatmap
sns.heatmap(gb, annot= True)
plt.title("Relationship Between Parent's Education & Student's Scores", fontsize=16, fontweight='bold')
plt.show()
```

```python
#From the above chart we can conclude that parent's edcuation have
#an impact on the students score.
```

```python
#To know how parents education affects the scores of the students
#we will use groupby fucntion.
gp = df.groupby("ParentMaritalStatus").agg({"MathScore":'mean', "ReadingScore":'mean',"WritingScore":'mean'})
gp
```

```python
#To analysis the above data using a heatmap
sns.heatmap(gp, annot= True)
plt.title("Relationship Between Parent's Marital Status & Student's Scores", fontsize=16, fontweight='bold')

plt.show()
```

```python
#From the above chart, it is seen that marital status of the parents 
#have very negligible impact on the student's scores.
```

```python
#To detect outliers we will plot boxplot
sns.boxplot(data = df, x ="MathScore")
plt.title("Detecting Outliers", fontsize=16, fontweight='bold')
plt.style.use('ggplot') 
plt.show()
```

```python
sns.boxplot(data = df, x ="ReadingScore")
plt.title("Detecting Outliers", fontsize=16, fontweight='bold')
plt.show()
```

```python
sns.boxplot(data = df, x ="WritingScore")
plt.title("Detecting Outliers", fontsize=16, fontweight='bold')
plt.show()
```

```python
#To calculate the unique values present in the EthnicGroup column
df["EthnicGroup"].unique()
```

```python
#Distribution of Ethnic Groups(percentage wise) using Pie chart
groupA = df.loc[(df["EthnicGroup"]== "group A")].count() #calculating values containing group A
groupB = df.loc[(df["EthnicGroup"]== "group B")].count() #calculating values containing group B
groupC = df.loc[(df["EthnicGroup"]== "group C")].count() #calculating values containing group C
groupD = df.loc[(df["EthnicGroup"]== "group D")].count() #calculating values containing group D
groupE = df.loc[(df["EthnicGroup"]== "group E")].count() #calculating values containing group E

print(groupA["EthnicGroup"])
print(groupB["EthnicGroup"])
print(groupC["EthnicGroup"])
print(groupD["EthnicGroup"])
print(groupE["EthnicGroup"])
```

```python
#Pie chart

l =["groupA", "groupB", "groupC", "groupD", "groupE"]
mlist =[groupA["EthnicGroup"],groupB["EthnicGroup"],
        groupC["EthnicGroup"],groupD["EthnicGroup"],
        groupE["EthnicGroup"]]
plt.pie(mlist, labels = l, autopct = "%1.2f%%")
plt.title("Distribution of Ethnic Group", fontsize=16, fontweight='bold',color="red")

plt.legend(fontsize ='small',bbox_to_anchor=(1.1, 0.5))
plt.style.use('ggplot') 
plt.show()
```
